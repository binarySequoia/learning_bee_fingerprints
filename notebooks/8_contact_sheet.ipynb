```python
%matplotlib inline
import random
from skimage import io
import matplotlib.pyplot as plt
from ipywidgets import interact
import numpy as np
import pandas as pd
import os
```

```python
train_df = pd.read_csv("train_id.csv")
test_df = pd.read_csv("test_id.csv")
```

```python
X_train_paths = list(train_df.X_train)
X_test_paths = list(test_df.X_test)
```

```python
X_train = np.array(io.imread_collection(X_train_paths))
X_test = np.array(io.imread_collection(X_test_paths))
```

```python
X_train.shape, X_test.shape
```

```python
y_train = np.array(train_df.y)
y_test = np.array(test_df.y)
```

```python
y_train.shape, y_test.shape
```

```python
fig=plt.figure(figsize=(30, 30))
columns = 20
rows = 10
j = 0


for i in range(10):
    idx = y_train[y_train == i]
    X_temp = X_train[idx]
    
    for x in X_temp:
        fig.add_subplot(rows, columns, j+1)
        plt.imshow(x)
        j += 1

        
plt.show()        
fig.savefig('train.pdf')   
```

```python
fig=plt.figure(figsize=(30, 30))
columns = 20
rows = 10
j = 0


for i in range(10):
    idx = y_test[y_test == i]
    X_temp = X_test[idx]
    
    for x in X_temp:
        fig.add_subplot(rows, columns, j+1)
        plt.imshow(x)
        j += 1

        
plt.show()        
fig.savefig('test.pdf') 
```
